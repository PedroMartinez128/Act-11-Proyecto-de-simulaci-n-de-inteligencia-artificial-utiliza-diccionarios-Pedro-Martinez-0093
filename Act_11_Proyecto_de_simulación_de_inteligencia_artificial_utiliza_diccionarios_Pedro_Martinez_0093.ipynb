#### Problema 38 Investigadores:
# ------Atributos-------
* Línea de investigación
* publicaciones científicas
* universidad
* becas obtenidas
* nivel de SNI
* laboratorio.


```python
# --- FUNCIONES DEL PROGRAMA ---

def mostrar_menu():
    """Muestra las opciones disponibles en la pantalla."""
    print("\n========================================")
    print("   MENÚ: SISTEMA DE INVESTIGADORES 🔬   ")
    print("========================================")
    print("1. Registrar nuevo investigador")
    print("2. Mostrar lista de investigadores")
    print("3. Salir")
    print("========================================")


def registrar_investigador(Investigadores):
    """Solicita los datos al usuario y los guarda en el diccionario investigadores."""
    print("\n--- REGISTRO DE NUEVO INVESTIGADOR ---")

    # Pedimos los datos para las llaves (Keys) y valores (Values)
    linea = input("Línea de investigación (ej. IA, Robótica): ")
    publicaciones = input("Número de publicaciones científicas: ")
    universidad = input("Universidad de procedencia: ")
    becas = input("Becas obtenidas (separadas por comas): ")
    sni = input("Nivel de SNI (Candidato, Nivel I, II, III): ")
    laboratorio = input("Nombre del laboratorio: ")

    # El diccionario individual con los nuevos atributos y sin el nombre
    investigadores = {
        "Línea de investigación": linea,
        "Publicaciones científicas": publicaciones,
        "Universidad": universidad,
        "Becas obtenidas": becas,
        "Nivel de SNI": sni,
        "Laboratorio": laboratorio
    }

    # Agregamos el diccionario 'investigadores' a la lista general 'Investigadores'
    Investigadores.append(investigadores)
    print(f"\n✅ ¡Investigador de la línea [{linea}] ha sido registrado con éxito!")


def mostrar_tabla(Investigadores):
    """Muestra los datos de la lista Investigadores en formato de tabla."""
    print("\n--- LISTADO DE INVESTIGADORES REGISTRADOS ---")

    # Si la lista está vacía, avisamos al usuario
    if len(Investigadores) == 0:
        print("Aún no hay investigadores registrados en el sistema.")
    else:
        # Encabezados de la tabla actualizados y alineados
       # Encabezados de la tabla más compactos y equilibrados
        print(f"{'Línea Inv.':<25} | {'Publ. Científicas':<20} | {'Universidad':<25} | {'Becas Obtenidas':<30} | {'Nivel SNI':<15} | {'Laboratorio':<30}")
        print("-" * 175)

        # Ciclo for para recorrer la lista y mostrar cada diccionario con sus nuevas claves
        for inv in Investigadores:
            print(f"{inv['Línea de investigación']:<25} | {inv['Publicaciones científicas']:<20} | {inv['Universidad']:<25} | {inv['Becas obtenidas']:<30} | {inv['Nivel de SNI']:<15} | {inv['Laboratorio']:<30}")

# --- PROGRAMA PRINCIPAL (MAIN) ---

# Al inicio, la lista contenedora empieza vacía
Investigadores = []

# Ciclo infinito para el menú interactivo
while True:
    mostrar_menu()
    opcion = input("Selecciona una opción (1-3): ")

    if opcion == "1":
        registrar_investigador(Investigadores)

    elif opcion == "2":
        mostrar_tabla(Investigadores)

    elif opcion == "3":
        print("\n¡Gracias por usar el Sistema de Investigadores! Saliendo... 🚀")
        break  # Rompe el ciclo while

    else:
        print("\n❌ Opción no válida. Por favor, elige 1, 2 o 3.")
```
